# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here
# Deleting the data doesnt seem to be the right option here because the outliers don't seem impossible, or to be cause by wrong input.
# Transformation seems the right choice in our case. 
# Due to the large discrepancies in termes of order of magnitude, a log transform would compress the values closer hence reducing the skewness.

# We use log1p because of the discontinuity on 0, since we have values at 0 in this set it is better to use log1p to avoid the discontinuity
data_clean = data.copy()
log_features = data_clean.drop(columns=['Channel','Region']).columns
log_features

data_clean[log_features] = np.log1p(data_clean[log_features])
```

```python
# plotting again on the cleaned data

columns = data_clean.drop(columns=['Region','Channel']).columns      
cols = 2
rows = int(len(columns) / cols)

fig, axes = plt.subplots(rows, cols, figsize=(14, 5 * rows))
axes = axes.flatten()

for i, feature in enumerate(columns):
# Histogram
    sns.histplot(data_clean[feature], kde=True, ax=axes[i])
    axes[i].set_title(f"{feature}")

# Create second axis sharing x-axis
    ax2 = axes[i].twinx()

# Boxplot (horizontal, placed above along y)
    sns.boxplot(
        x=data_clean[feature],
        ax=ax2,
        width=0.2,
        color = 'orange',
        boxprops=dict(alpha=0.5),  # transparency
    )

# Hide second y axis
    ax2.set_yticks([])
    ax2.set_ylabel("")

# Title
    #plt.title(f"Distribution of {feature}")

plt.show()
```

```python
# Checking skewnees evolution
for column in columns:
    print(f'Skewness of {column}: Raw: {data[column].skew():.2f} | Log: {data_clean[column].skew():.2f}')
```

**Your comment here**

-  The transformation using the log function strongly affected the distribution of the data in terms of scaling and skewness.
-  All of the strong skewness coefficients on raw data were above 1. After the log transformation these coefficients have dropped below 0 showing a left skewing of the data. 
- The coefficient have also drastically decreased in termes of absolute value. Now 'Milk','Grocery', 'Frozen' and 'Detergents_Paper' have a coefficient below 1 (in absolute value) which translate that they are more symmetrical now.
- The same can be said for 'Fresh' and 'Delicassen' for which the value also decreased. The absolute value still remains above 1 which shows that these features were over-corrected byt the transformation.

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:

# We are going to try different number of clusters from 1 to 100 in the KMeans method:
# We will use the .inertia_ to assess the quality of the clusters in each case
wcss = []
for i in range(1,100):
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(customers_scale)
    wcss.append(kmeans.inertia_)    # Lower wcss/inertia = compact cluster = Good // Higher wcss/inertia = loose cluster = Bad
wcss

# we plot the results to determine the optimal number of clusters
plt.plot(range(1,100),wcss,marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

**Your comment here**

- The increase of the number of clusters decrease the WCSS (or inertia) coefficient translating clusters becoming more compact.
- With a range of up to 100 clusters we can see that the WCSS decreases slower at high cluster number and tends to be converging. However increasing the number of cluster also means that we reduce the size of each cluster. therefore For a set of 440 data point having more than 100 clusters means that each cluster on average will not have more than 4 values. This can lead to complications.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here

# Increasing eps to 0.7
db2 = DBSCAN(eps=0.7)
data_clean['labels_DBSCAN_07'] = db2.fit(customers_scale).labels_

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(x = data_clean['Frozen'],y = data_clean['Delicassen'],hue = data_clean['labels_DBSCAN'],ax=axes[0])
axes[0].set_title('Frozen vs Delicassen (eps=0.5)')
sns.scatterplot(x = data_clean['Frozen'],y = data_clean['Delicassen'],hue = data_clean['labels_DBSCAN_07'], ax=axes[1])
axes[1].set_title('Frozen vs Delicassen (eps=0.7)')
plt.tight_layout()
plt.show()

print(f'eps = 0.5: {data_clean['labels_DBSCAN'].value_counts()} | eps = 0.7: {data_clean['labels_DBSCAN_07'].value_counts()}')
# Increasing the eps value, increase the radius at which the samples are considered in the neighborhood of the other.
# The algorithm starts populated and creating more clusters when incresing the value
```

```python
# Changing min_samples to 3
db2 = DBSCAN(eps=0.5,min_samples=3)
data_clean['labels_DBSCAN_3'] = db2.fit(customers_scale).labels_

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(x = data_clean['Frozen'],y = data_clean['Delicassen'],hue = data_clean['labels_DBSCAN'],ax=axes[0])
axes[0].set_title('Frozen vs Delicassen (min_samples=5)')
sns.scatterplot(x = data_clean['Frozen'],y = data_clean['Delicassen'],hue = data_clean['labels_DBSCAN_3'], ax=axes[1])
axes[1].set_title('Frozen vs Delicassen (min_samples=3)')
plt.tight_layout()
plt.show()

print(f'N = 5: {data_clean['labels_DBSCAN'].value_counts()} | N = 3: {data_clean['labels_DBSCAN_3'].value_counts()}')
# Decreasing the min_sample value facilitates the creation of clusters.
# Reducing the value to 3 allowed the algorithm to create more clusters, up to 7. 
```

**Your comment here**
- Increasing the eps value, increase the radius at which the samples are considered in the neighborhood of the other.
- The algorithm starts populated and creating more clusters when incresing the value
- Decreasing the min_sample value facilitates the creation of clusters.
- Reducing the value to 3 allowed the algorithm to create more clusters, up to 7.
